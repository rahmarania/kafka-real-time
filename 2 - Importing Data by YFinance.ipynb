### Stock Data Case 

#### Fetching data from API YFinance
about yfinance 
- update about currency, scrap for finance data from yahoo finance
- this API has 1 minute delayed, this API can't be used for real-time
- most used parameter is Tickers, Period, and Interval
- download() fetching data, has parameters:

  - **Tickers (STR or List)** <br>
    The symbol of the stock or asset that you want to retrieve the data
    Example: "BTC-USD" or ["BTC-USD", "ETH-USD"]
    
  - **Period (str)** <br>
    The time range of the data you want to retrieve.
    Examples: "1mo" (1 month), "5d" (5 days), "1y" (1 year)
    Another option: "max" (maximum historical data available)  
    
  - **interval (str)** <br>
    The time interval between data retrieved.
    Examples: "1d" (daily), "5m" (5 minutes), "1h" (1 hour)
    
  - Start and End (STR, optional) <br>
    Used if you want to retrieve data within a specific time rnge.
    
  - Format: "YYYY-MM-DD" <br>
    Example: start="2023-01-01", end="2023-06-01"
    
  - auto_adjust (bool, default: True) <br>
    If True, the price will be automatically adjusted to the dividend distribution and stock split.
    
  - actions (bool, default: False) <br>
    If True, it will include information about dividends and stock splits.
    
  - threads (bool, default: True) <br>
    If True, use multi-threading to speed up data retrieval.

```python
import pandas as pd
import yfinance as yf
```

```python
# tickers because we want to use more than 1 ticker (bitcoin and etherium)
# period 1 month
# interval 5 minute
dt = yf.download(tickers = ["BTC-USD","ETH-USD"],
           period = "1mo",
           interval = "5m") 
dt
```

```python
# show close and volume
dt[['Close','Volume']]
```

### .xs()
- Cross Section function is used for data that has multiple level, in this case, our data has 2 levels.
- This function has 3 parameters
  - **key** what column we need
  - **level** what level of the column we want to take
  - **axis** 1 for column, 0 for row

```python
# take only btc data, create dt_btc
dt_btc = dt.xs("BTC-USD", level='Ticker',axis=1)
dt_btc
```

```python
# reset the index because we want to do data processing by column
dt_btc = dt_btc.reset_index()
dt_btc
```

### Groupby
In case we want to show data based on category or something, we can use groupby

```python
# grouping by date
# show the movement per day
dt_btc['Date'] = dt_btc['Datetime'].dt.date

# aggregate mean close price per day 
# we don't know exact how much mean prie in that day
dt_avg_dailyPrice = dt_btc.groupby('Date')['Close'].mean()
dt_avg_dailyPrice
```

```python
# if we want to use multiple calculation for multiple column, we can use agg()
dt_btc.groupby('Date').agg({
    'Close':'mean',
    'Volume':'std'
})
```

### Data Visualization


```python
# visualization using basic python
dt_avg_dailyPrice.plot()
```
